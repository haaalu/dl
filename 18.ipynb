```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
```

```python
file_path = 'DL_LAB_EXAM/DL_LAB_EXAM/Datasets/UCI_letter_recognition/letter-recognition.data'
column_names = ['letter'] + [f'feature_{i}' for i in range(1, 17)]
data = pd.read_csv(file_path, header=None, names=column_names)
```

```python
X = data.iloc[:, 1:].values  # Features
y = data.iloc[:, 0].values 
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_categorical, test_size=0.2, random_state=42)
```

```python
print("Training model with batch size 32...")
model_32 = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(26, activation='softmax')
])

model_32.compile(optimizer=Adam(learning_rate=0.001),
                loss='categorical_crossentropy',
                metrics=['accuracy'])

history_32 = model_32.fit(X_train, y_train,
                         batch_size=32,
                         epochs=20,
                         validation_data=(X_test, y_test),
                         verbose=1)
```

```python
print("\nTraining model with batch size 64...")
model_64 = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(26, activation='softmax')
])

model_64.compile(optimizer=Adam(learning_rate=0.001),
                loss='categorical_crossentropy',
                metrics=['accuracy'])

history_64 = model_64.fit(X_train, y_train,
                         batch_size=64,
                         epochs=20,
                         validation_data=(X_test, y_test),
                         verbose=1)
```

```python
# Plot comparison
plt.figure(figsize=(12, 5))

# Loss comparison
plt.subplot(1, 2, 1)
plt.plot(history_32.history['val_loss'], label='32-batch val loss')
plt.plot(history_64.history['val_loss'], label='64-batch val loss')
plt.title('Validation Loss Comparison')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

# Accuracy comparison
plt.subplot(1, 2, 2)
plt.plot(history_32.history['val_accuracy'], label='32-batch val acc')
plt.plot(history_64.history['val_accuracy'], label='64-batch val acc')
plt.title('Validation Accuracy Comparison')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
print("\nEvaluation Results:")

# Batch size 32 results
train_loss_32, train_acc_32 = model_32.evaluate(X_train, y_train, verbose=0)
test_loss_32, test_acc_32 = model_32.evaluate(X_test, y_test, verbose=0)
print("\nBatch size 32:")
print(f"Training Accuracy: {train_acc_32:.4f}")
print(f"Testing Accuracy: {test_acc_32:.4f}")

# Batch size 64 results
train_loss_64, train_acc_64 = model_64.evaluate(X_train, y_train, verbose=0)
test_loss_64, test_acc_64 = model_64.evaluate(X_test, y_test, verbose=0)
print("\nBatch size 64:")
print(f"Training Accuracy: {train_acc_64:.4f}")
print(f"Testing Accuracy: {test_acc_64:.4f}")
```

```python
sample_indices = np.random.choice(len(X_test), 5, replace=False)
sample_data = X_test[sample_indices]
true_labels = np.argmax(y_test[sample_indices], axis=1)

print("\nSample Predictions:")
print("True\t32-batch\t64-batch")
for i in range(5):
    pred_32 = np.argmax(model_32.predict(sample_data[i:i+1]))
    pred_64 = np.argmax(model_64.predict(sample_data[i:i+1]))
    print(f"{label_encoder.inverse_transform([true_labels[i]])[0]}\t"
          f"{label_encoder.inverse_transform([pred_32])[0]}\t\t"
          f"{label_encoder.inverse_transform([pred_64])[0]}")
```
